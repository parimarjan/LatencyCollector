```python
import numpy as np
import matplotlib.pyplot as plt

N = 10000  # Number of machines
K = 10000    # Number of samples per machine

# Step 1: Generate the samples
samples = np.random.uniform(0, 1, (N, K))

# Step 2: Calculate CDFs
# Since the distribution is uniform, the CDF at a value x is just x
# For other distributions, you could use np.sort() and calculate the CDF manually

# Step 3: Calculate percentiles
p50 = np.percentile(samples, 50, axis=1)  # 50th percentile (median) for each machine
p75 = np.percentile(samples, 75, axis=1)  # 75th percentile for each machine

# Step 4: Plot the CDFs
# Plot CDFs for a subset of machines to avoid overplotting
for i in range(min(N, 50)):  # Let's plot CDFs for 50 machines
    sorted_samples = np.sort(samples[i])
    plt.step(sorted_samples, np.linspace(0, 1, K, endpoint=False), label=f'Machine {i+1}' if i < 10 else '')

plt.xlabel('Value')
plt.ylabel('CDF')
plt.title('CDFs of Uniform Samples Across Machines')
plt.legend()
plt.show()

# Step 5: Plot the distribution of percentiles across machines
plt.figure()

# Plot histogram for the 50th percentile
plt.hist(p50, bins=50, alpha=0.5, label='50th Percentile')

# Plot histogram for the 75th percentile
plt.hist(p75, bins=50, alpha=0.5, label='75th Percentile')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of Percentiles Across Machines')
plt.legend()
plt.show()
```
